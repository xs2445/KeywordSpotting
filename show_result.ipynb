```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
log_folder = './logs/2022-04-30'
val_acc_list = []
train_acc_list = []
dataset_list = []
for root, _, files in os.walk(log_folder):
    if root == log_folder:
        continue
    dataset_list.append(int(root.split('-')[-1]))
    val_acc_list.append(np.load(os.path.join(root, 'val_acc_hist.npy')))
    train_acc_list.append(np.load(os.path.join(root, 'train_acc_hist.npy')))

val_acc_list = [x for _,x in sorted(zip(dataset_list,val_acc_list))]
train_acc_list = [x for _,x in sorted(zip(dataset_list,train_acc_list))]
dataset_list = sorted(dataset_list)
```

```python
plt.figure()
plt.plot([np.max(x) for x in val_acc_list])
plt.title('val_acc w.r.t. f_c_max (res15-n32-q3-a1-fcmin100)')
plt.xlabel('f_c_max')
plt.ylabel('validation accuracy')
plt.xticks(range(len(dataset_list)), dataset_list, rotation=70)
plt.grid()
plt.savefig(os.path.join(log_folder, 'result.jpg'))
plt.show()
```

```python
plt.figure()
plt.plot([np.max(x) for x in train_acc_list])
plt.title('train_acc w.r.t. f_c_max (res15-n32-q3-a1-fcmin100)')
plt.xlabel('f_c_max')
plt.ylabel('training accuracy')
plt.xticks(range(len(dataset_list)), dataset_list, rotation=70)
plt.grid()
plt.show()
```
