```python
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

q 2022-05-06 2022-05-07

```python
log_folder = ['./logs/2022-05-06', './logs/2022-05-07']
val_acc_list = []
train_acc_list = []
dataset_list = []
for log_f in log_folder:
    for root, _, files in os.walk(log_f):
        if root == log_f:
            continue
        dataset_list.append(float(root.split('-')[-4][1:]))
        val_acc_list.append(np.load(os.path.join(root, 'val_acc_hist.npy')))
        train_acc_list.append(np.load(os.path.join(root, 'train_acc_hist.npy')))

print(dataset_list)

val_acc_list = [x for _,x in sorted(zip(dataset_list,val_acc_list))]
train_acc_list = [x for _,x in sorted(zip(dataset_list,train_acc_list))]
dataset_list = sorted(dataset_list)
```

```python
plt.figure()
# plt.plot([np.max(x) for x in val_acc_list])
plt.plot(dataset_list, [np.max(x) for x in val_acc_list])
plt.title('val_acc w.r.t. q (res15-n32-a1-fcmin100-fcmax4e3)')
plt.xlabel('q')
plt.ylabel('validation accuracy')
# [0.2, 0.4, 0.6, 0.8, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 6, 7, 8, 9, 10, 15, 20, 30]
selected = [dataset_list[x] for x in [0,6,10,13,15,17,18,19,-1]]
print()
plt.xticks(selected, selected, rotation=70)
plt.grid()
plt.savefig(os.path.join(log_folder[1], 'result.jpg'))
plt.show()
print([np.max(x) for x in val_acc_list])
```

```python
plt.figure()
# plt.plot([np.max(x) for x in train_acc_list])
plt.plot(dataset_list, [np.max(x) for x in train_acc_list])
plt.title('train_acc w.r.t. q (res15-n32-a1-fcmin100-fcmax4e3)')
plt.xlabel('q')
plt.ylabel('training accuracy')
# plt.xticks(range(len(dataset_list)), dataset_list, rotation=70)
plt.xticks(selected, selected, rotation=70)
plt.grid()
plt.show()
```
